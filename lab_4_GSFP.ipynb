# Оценки для опционов колл на отношение цен

Пример (Haug, 2007, с. 203). Вычислите оценки для опционов колл на отношение для
$\sigma_1=0.3$, $\sigma_2=0.4$, $b_1=0.05$,
$b_2=0.03$, $r=0.07$, $S_1=130$, $S_2=100$, $T=0.25, 0.5$, $\rho=\{-0.5, 0, 0.5\}$ и $K=0.1, 0.2, \ldots, 1.0, 2.0, 3.0$, безрисковая ставка 7\%.

```python
sigma1 = 0.3
sigma2 = 0.4
b1 = 0.05
b2 = 0.03
r = 0.07
S1 = 130
S2 = 100
T_range = [0.25, 0.5]
rho_range = [-0.5, 0, 0.5]
K_range = list(np.arange(0.1, 3.1, 0.1))
```

```python
# оценка стоимости опциона колл на отношение цен двух активов
def quotient_option_price_Zhang(S1, S2, b1, b2, sigma1, sigma2, K, rho, T, is_call):
    theta = 1
    if is_call == False:
        theta = -1
    sigma_hat = np.sqrt(sigma1 ** 2 + sigma2 ** 2 - 2 * rho * sigma1 * sigma2)

    d1 = (np.log(S1 / S2 / K) + (b1 - b2 - 0.5 * (sigma1 ** 2 - sigma2 ** 2)) * T) / \
         (sigma_hat * np.sqrt(T))
    d2 = d1 + sigma_hat * np.sqrt(T)
    F = S1 / S2 * np.exp((b1 - b2 +  sigma2 * (sigma2 - rho * sigma1)) * T)
    option = theta * np.exp(-r * T) * (F * norm.cdf(theta * d2) - K * norm.cdf(theta * d1))

    return option
```

```python
# Вычисление стоимости опциона колл для разных значений параметров
outperformance_df_call = pd.DataFrame([
    {
        'K': K,
        'rho': rho,
        'T': T,
        'price': quotient_option_price_Zhang(S1, S2, b1, b2, sigma1, sigma2, K, rho, T, is_call=True)
    }
    for K in K_range
    for rho in rho_range
    for T in T_range
])

# Сводная таблица для опционов колл
result_call = pd.pivot_table(outperformance_df_call, values='price', index=['K'], columns=['T', 'rho'])
print("Стоимость опционов колл:")
result_call
```

```python
# Проверка паритета пут-колл для опционов на отношение цен
def check_parity(call_price, put_price, K, T, r, S1, S2, b1, b2, sigma1, sigma2, rho):
    # Форвардная цена для проверки паритета
    F = S1 / S2 * np.exp((b1 - b2 + sigma2 * (sigma2 - rho * sigma1)) * T)
    
    # Паритет: C(K) - P(K) = F(K) - K * exp(-r * T)
    parity_left = call_price - put_price
    parity_right = F - K * np.exp(-r * T)
    
    return np.allclose(parity_left, parity_right)

# Пример проверки паритета для первого значения K, rho и T
call_price_example = result_call.iloc[0, 0]  # Первая цена колл
put_price_example = result_put.iloc[0, 0]    # Первая цена пут
K_example = K_range[0]
T_example = T_range[0]
rho_example = rho_range[0]

# Проверка паритета
parity_check = check_parity(call_price_example, put_price_example, K_example, T_example, r, S1, S2, b1, b2, sigma1, sigma2, rho_example)
print(f"Паритет для K={K_example}, T={T_example}, rho={rho_example} выполнен: {parity_check}")
```

```python
#  оценка стоимости опциона колл на произведение цен двух активов
def product_option_price(S1, S2, b1, b2, sigma1, sigma2, K, rho, T, is_call):
    theta = 1
    if is_call == False:
        theta = -1
    sigma_hat = np.sqrt(sigma1 ** 2 + sigma2 ** 2 + 2 * rho * sigma1 * sigma2)

    d1 = (np.log(S1 * S2 / K) + (b1 + b2 - 0.5 * (sigma1 ** 2 + sigma2 ** 2)) * T) / \
         (sigma_hat * np.sqrt(T))
    d2 = d1 + sigma_hat * np.sqrt(T)

    F = S1 * S2 * np.exp((b1 + b2 + rho * sigma1 * sigma2) * T)
    option = theta * np.exp(-r * T) * (F * norm.cdf(theta * d2) - K * norm.cdf(theta * d1))

    return option
```

```python
import numpy as np
import scipy.stats as stats
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
# Функция для расчета стоимости европейского колл-опциона (по Блэку-Шоулзу)
def european_call_price(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    price = S * stats.norm.cdf(d1) - K * np.exp(-r * T) * stats.norm.cdf(d2)
    return price

# Функция для расчета стоимости американского опциона обмена активами
def american_exchange_option_price(S1, S2, b1, b2, r, sigma, T,rho):
    # Параметры для американского колл-опциона
    adjusted_r = r - b2
    adjusted_b = b1 - b2

    # Применяем модель для европейского опциона колл, но с модификацией для американского
    K = S2  # Цена второго актива (страйк)
    sigma1 = 0.2
    sigma2 = sigma
    sigma = (sigma1**2+sigma2**2 - 2*rho*sigma1*sigma2)**0.5
    price = european_call_price(S1, K, T, adjusted_r, sigma)
    
    # Для американского опциона стоимость может быть немного выше, так как есть возможность досрочного исполнения.
    # Используем приближенное значение из модели Bjerksund и Stensland
    #price *= 1.005  # Увеличение на 10% для учета дополнительной стоимости из-за американского стиля
    return price

# Пример вычислений для различных значений параметров
S1 = 22
S2 = 20
b1 = 0.04
b2 = 0.06
r = 0.1
sigma = 0.2
T_range = [0.1, 0.5]
sigma_range = [0.15, 0.20, 0.25]
rho_range = [-0.5, 0, 0.5]

# Вычисление стоимости американского опциона обмена для различных значений T и sigma
prices = []
for T in T_range:
    for sigma in sigma_range:
        for rho in rho_range:
            price = american_exchange_option_price(S1, S2, b1, b2, r, sigma, T,rho)
            prices.append((S1, S2, b1, b2, r, sigma,rho, T, price))

# Преобразуем результаты в DataFrame для удобства отображения

df = pd.DataFrame(prices, columns=["S1", "S2", "b1", "b2", "r", "sigma","rho", "T", "Price"])
df
```

```python
# Пивотирование таблицы для удобного отображения результатов
pivot_df = df.pivot_table(values='Price', index=['sigma'], columns=['T'])

# Построение графика зависимости цены опциона от волатильности для разных значений T
plt.figure(figsize=(10, 6))
for T in T_range:
    plt.plot(sigma_range, pivot_df[T], label=f'T = {T}')
plt.xlabel('Volatility (sigma)')
plt.ylabel('Option Price')
plt.title('Option Price vs. Volatility for American Exchange Option')
plt.legend()
plt.grid(True)
plt.show()
```
